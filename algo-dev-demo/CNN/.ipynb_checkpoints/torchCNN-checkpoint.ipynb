```python
import numpy as np
import torch
import torchvision
import torchvision.transforms as transforms
%matplotlib inline
```

```python
#Seed the random generator for consistent results
seed = 42
np.random.seed(seed)
torch.manual_seed(seed)
```

Pytorch has functionality for defining a subset of a dataset for computational efficiency.

```python
from torch.utils.data.sampler import SubsetRandomSampler

#Training
n_training_samples = 20000
train_sampler = SubsetRandomSampler(np.arange(n_training_samples, dtype=np.int64))

#Validation
n_val_samples = 5000
val_sampler = SubsetRandomSampler(np.arange(n_training_samples, n_training_samples + n_val_samples, dtype=np.int64))

#Test
n_test_samples = 5000
test_sampler = SubsetRandomSampler(np.arange(n_test_samples, dtype=np.int64))
```

Pytorch's CIFAR-10 dataset is in PIL (Python Image Library) format, and we need to convert them into Tensors.

```python
num_workers = 2
test_batch_size = 4

#The compose function allows for multiple transforms
#transforms.ToTensor() converts our PILImage to a tensor of shape (C x H x W) in the range [0,1]
#transforms.Normalize(mean,std) normalizes a tensor to a (mean, std) for (R, G, B)
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

train_set = torchvision.datasets.CIFAR10(root='./cifardata', train=True, download=True, transform=transform)

test_set = torchvision.datasets.CIFAR10(root='./cifardata', train=False, download=True, transform=transform)
```

Design and set the labels for the CIFAR-10 dataset.

```python
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

Define the Convolutional Neural Network:

```python
from torch.autograd import Variable
import torch.nn.functional as F
```

Elements of the CNN that Torch provides functions for:

* **nn.Conv2d**(in_channels, out_channels, kernel_size, stride, padding)
* **nn.MaxPool2d**(kernel_size, stride, padding)
* **F.relu**(Z1)
* **x.view**(new_shape) – similar to np.reshape
* **nn.Linear**(in_features, out_features) – applies a linear transformation to the incoming data (fully connected layer)

Here we go!

```python
from torch.autograd import Variable
import torch.nn.functional as F

class SimpleCNN(torch.nn.Module):
    
    #Our batch shape for input x is (3, 32, 32)
    
    def __init__(self):
        super(SimpleCNN, self).__init__()
        
        #Input channels = 3, output channels = 18
        self.conv1 = torch.nn.Conv2d(3, 18, kernel_size=3, stride=1, padding=1)
        self.pool = torch.nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        
        #4608 input features, 64 output feature (see sizing flow below)
        self.fc1 = torch.nn.Linear(18 * 16 * 16, 64)
        
        #64 input feature, 10 output features for our 10 defined classes
        self.fc2 = torch.nn.Linear(64, 10)
        
    def forward(self, x):
        
        #Computes the activation of the first convolution
        #Size changes from (3, 32, 32) to (18, 32, 32)
        x = F.relu(self.conv1(x))
        
        #Pooling descales the input by a factor of the kernel_size (2)
        #Size changes from (18, 32, 32) to (18, 16, 16)
        x = self.pool(x)
        
        #Reshape data to input to the input layer of the neural net
        #Size changes from (18, 16, 16) to (1, 4608)
        #Recall that the -1 infers this dimension from the other given dimension
        x = x.view(-1, 18 * 16 *16)
        
        #Computes the activation of the first fully connected layer
        #Size changes from (1, 4608) to (1, 64)
        x = F.relu(self.fc1(x))
        
        #Computes the second fully connected layer (activation applied later)
        #Size changes from (1, 64) to (1, 10)
        x = self.fc2(x)
        return(x)
```

```python
class LinearClassifier(torch.nn.Module):
    
    def __init__(self):
        super(LinearClassifier, self).__init__()
        self.linear = torch.nn.Linear(32 * 32 * 3, 10)
        
    def forward(self, x):
        #Flatten input from (3, 32, 32) to (1, 3072)
        x = x.view(x.size(0), -1)
        return(self.linear(x))
```

Define a loss function and optimizer:

```python
import torch.optim as optim

def createLossAndOptimizer(net, learning_rate=0.001):
    
    #Loss function
    loss = torch.nn.CrossEntropyLoss()
    
    #Optimizer
    optimizer = optim.Adam(net.parameters(), lr=learning_rate)
    
    return(loss, optimizer)
```

Create data loaders to iterate through our data and pass to the net:

```python
#DataLoader takes in a dataset and a sampler for loading (num_workers deals with system level memory) 
def get_train_loader(batch_size):
    train_loader = torch.utils.data.DataLoader(train_set, batch_size=test_batch_size,
                                           sampler=train_sampler, num_workers=2)
    return(train_loader)
```

```python
#Test and validation loaders have constant batch sizes, so we can define them directly
test_loader = torch.utils.data.DataLoader(test_set, batch_size=4, sampler=test_sampler, num_workers=2)
val_loader = torch.utils.data.DataLoader(train_set, batch_size=128, sampler=val_sampler, num_workers=2)
```

#### Training loop

```python
import time

def trainNet(net, batch_size, n_epochs, learning_rate):
    
    #Print all of the hyperparameters of the training iteration:
    print("===== HYPERPARAMETERS =====")
    print("batch_size=", batch_size)
    print("epochs=", n_epochs)
    print("learning_rate=", learning_rate)
    print("=" * 30)
    
    #Get training data
    train_loader = get_train_loader(batch_size)
    n_batches = len(train_loader)
    
    #Create our loss and optimizer functions
    loss, optimizer = createLossAndOptimizer(net, learning_rate)
    
    #Time for printing
    training_start_time = time.time()
    
    #Loop for n_epochs
    for epoch in range(n_epochs):
        
        running_loss = 0.0
        print_every = n_batches // 10
        start_time = time.time()
        total_train_loss = 0
        
        for i, data in enumerate(train_loader, 0):
            
            #Get inputs
            inputs, labels = data
            
            #Wrap them in a Variable object
            inputs, labels = Variable(inputs), Variable(labels)
            
            #Set the parameter gradients to zero
            optimizer.zero_grad()
            
            #Forward pass, backward pass, optimize
            outputs = net(inputs)
            loss_size = loss(outputs, labels)
            loss_size.backward()
            optimizer.step()
            
            #Print statistics
            running_loss += loss_size.data[0]
            total_train_loss += loss_size.data[0]
            
            #Print every 10th batch of a an epoch
            if (i + 1) % (print_every + 1) == 0:
                print("Epoch {}, {:d}% \t train_loss: {:.2f} took: {:.2f}s".format(
                        epoch+1, int(100 * (i+1) / n_batches), running_loss / print_every, time.time() - start_time))
                #Reset running loss and time
                running_loss = 0.0
                start_time = time.time()
            
        #At the end of the epoch, do a pass on the validation set
        total_val_loss = 0
        for inputs, labels in val_loader:
            
            #Wrap tensors in Variables
            inputs, labels = Variable(inputs), Variable(labels)
            
            #Forward pass
            val_outputs = net(inputs)
            val_loss_size = loss(val_outputs, labels)
            total_val_loss += val_loss_size.data[0]
            
        print("Validation loss = {:.2f}".format(total_val_loss / len(val_loader)))
        
        #Save checkpoint
        state = {'epoch': epoch, 'model_state': net.state_dict(), 'optimizer': optimizer.state_dict()}
        torch.save(state, 'checkpoint.pth.tar')
        
    print("Training finished, took {:.2f}s".format(time.time() - training_start_time))
```

```python
CNN = SimpleCNN()
trainNet(CNN, batch_size=32, n_epochs=2, learning_rate=0.001)
```

```python
from PIL import Image

def preprocess(input_image):
    #Normalize and resize image
    normalize = transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    composed = transforms.Compose([transforms.Resize(32), transforms.ToTensor(),normalize])
    img = Image.open(input_image)
    img.load()
    output = composed(img)
    return output
    
def predict(input, newCNN):
    outputs = newCNN(Variable(torch.stack([input])))
    _, predicted = torch.max(outputs.data, 1)
    return int(predicted.numpy())
```

#### Testing accuracy

```python
def datasetAccuracy(net, data_loader, name=""):
    correct = 0
    total = 0
    for data in data_loader:
        inputs, labels = data
        outputs = net(Variable(inputs))
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum()
    accuracy = 100 * correct / total
    print('Accuracy of the network on the {} {} images: {:.2f} %'.format(total, name, accuracy))
```

```python
#Training accuracy
datasetAccuracy(newCNN, get_train_loader(batch_size=4), "train")
```

```python
#Validation accuracy
datasetAccuracy(newCNN, val_loader, "validation")
```

```python
#Test accuracy
datasetAccuracy(CNN, test_loader, "test")
```

#### Accuracy per class

```python
def classAccuracy(net):
    
    n_classes = 10
    #Initialize empty lists of 0.0 of length n_classes
    class_correct = list(0.0 for i in range(n_classes))
    class_total = list(0.0 for i in range(n_classes))
    
    #Initialize empty (n_classes, n_classes) matrix to act as confusion matrix
    confusion_matrix = np.zeros((n_classes, n_classes), dtype=np.int64)
    
    #Iterate through test data
    for data in test_loader:
        
        #Grab inputs and run net to get outputs
        inputs, labels = data
        outputs = net(Variable(inputs))
        
        #Grab predictions by taking the max of the output layer (index 0 is the actual probabilities)
        predicted = torch.max(outputs.data, 1)[1]
        #The np.squeeze() function gets rid of extra dimensions in np arrays
        c = (predicted == labels).squeeze()
        
        #The number of labels and predictions should equal the batch_size
        for i in range(test_batch_size):
            
            #Set (label, prediction) in the CM to 1 – correct classifications will always be diagonal
            confusion_matrix[labels[i], predicted[i]] += 1
            
            #Add whether a label was correct of not to the class_correct list
            label = labels[i]
            class_correct[label] += c[i] #If the prediction was incorrect, c[i] is 0
            
            #Increment the total number of times the class appears in the class_total list
            class_total[label] += 1
            
            
    print("{:<10} {:^10}".format("Class", "Accuracy (%)"))
    for i in range(n_classes):
        print('{:<10} {:^10.2f}'.format(
            classes[i], 100 * class_correct[i] / class_total[i]))
    return confusion_matrix
```

### Improvements

Let's add more layers and fool with the hyperparameters!

A super important part of designing the SimpleCNN class is using the right sizes for each layer. Using convolutions and pooling can make that difficult, so here's the formula for determining output size given the parameters of preprocessing:

```python
def outputSize(in_size, kernel_size, stride, padding):
    output = int((in_size - kernel_size + 2*(padding)) / stride) + 1
    return(output)
```

Simple addition: let's try adding a second layer of convolution and pooling.

```python
class MyCNN(torch.nn.Module):
    
    def __init__(self):
        super(MyCNN, self).__init__()
        
        #Data comes in as (3, 32, 32)
        self.conv1 = torch.nn.Conv2d(3, 6, kernel_size=3, stride=1, padding=1) #--> (10, 15, 15)
        self.conv2 = torch.nn.Conv2d(6, 12, kernel_size=3, stride=1, padding=1) #--> (18, 13, 13)
        self.conv3 = torch.nn.Conv2d(12, 18, kernel_size=3, stride=1, padding=1) #--> (18, 13, 13)
        
        self.pool1 = torch.nn.MaxPool2d(kernel_size=2, stride=1, padding=0) #--> (10, 14, 14)

        #Fully connected layers
        self.fc1 = torch.nn.Linear(18 * 6 * 6, 64) #--> (1, 64)
        self.fc2 = torch.nn.Linear(64, 10)
        
    def forward(self, x):
        
        x = F.relu(self.conv1(x))
        x = self.pool1(x)
        x = F.relu(self.conv2(x))
        x = self.pool2(x)
        x = x.view(-1, 18 * 6 * 6)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return(x)
```
